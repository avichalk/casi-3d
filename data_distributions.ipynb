```python
% matplotlib inline

from preprocessing import load_fits, density_preprocessing, normalize
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
co_x, co_classify_y = load_fits(['../data/CO_integrate_binarized/co_all_rot.fits',
                                 '../data/CO_integrate_binarized/co_all_rot_tracer.fits'])

co_x, co_regress_y = load_fits(['../data/CO_integrate/co_all_rot.fits',
                                '../data/CO_integrate/co_all_rot_tracer.fits'])

density_x, density_y = load_fits(['../data/Density/wind_2cr_flatrho_2315_256.fits',
                                  '../data/Density/wind_2cr_flattracer_2315_256.fits'])
```

# <sup>12</sup>CO Investigation
## Raw Values

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 8))

sns.distplot(co_x.flatten(), kde=False, ax=axes[0])
sns.distplot(co_regress_y.flatten(), kde=False, ax=axes[1])
sns.distplot(co_classify_y.flatten(), kde=False, ax=axes[2])
axes[0].set_title('$^{12}$CO')
axes[1].set_title('$^{12}$CO Tracer')
axes[2].set_title('$^{12}$CO Tracer Binarized');
```

## Normalized

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

sns.distplot(normalize(co_x.flatten()), kde=False, ax=axes[0])
sns.distplot(normalize(co_regress_y.flatten()), kde=False, ax=axes[1])
axes[0].set_title('Normalized $^{12}$CO')
axes[1].set_title('Normalized $^{12}$CO Tracer');
```

## log<sub>10</sub>( 1 + x - min(x) )

```python
log_op = lambda x: np.log10(1. + x.flatten() - np.min(x))

fig, axes = plt.subplots(1, 2, figsize=(20, 8))

sns.distplot(log_op(co_x), kde=False, ax=axes[0])
sns.distplot(log_op(co_regress_y), kde=False, ax=axes[1])
axes[0].set_title('$\log_{10}( 1 + ^{12}$CO)')
axes[1].set_title('$\log_{10}( 1 + ^{12}$CO Tracer)');
```

# Density Investigation

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

sns.distplot(density_x.flatten(), kde=False, ax=axes[0])
sns.distplot(density_y.flatten(), kde=False, ax=axes[1])
axes[0].set_title('Density')
axes[1].set_title('Target');
```
